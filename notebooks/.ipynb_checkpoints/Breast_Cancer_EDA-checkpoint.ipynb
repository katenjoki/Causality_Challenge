<h1>Exploratory Data Analysis of Breast Cancer Data</h1>

* Features are computed from a digitized image of a fine needle aspirate (FNA) of a breast mass. They describe characteristics of the cell nuclei present in the image.

<h2>Data Attributes</h2>

* ID number
* Diagnosis (M = malignant, B = benign)

**Ten real-valued features are computed for each cell nucleus:**

* radius (mean of distances from center to points on the perimeter)
* texture (standard deviation of gray-scale values)
* perimeter
* area
* smoothness (local variation in radius lengths)
* compactness (perimeter^2 / area - 1.0)
* concavity (severity of concave portions of the contour)
* concave points (number of concave portions of the contour)
* symmetry
* fractal dimension ("coastline approximation" - 1)

The mean, standard error and "worst" or largest (mean of the three largest values) of these features were computed for each image

Data Visualization

```python
M = df[(df['diagnosis'] != 0)]
B = df[(df['diagnosis'] == 0)]

#Count plot
def plot_count(data:pd.DataFrame,col:str)->None:
    sns.countplot(data[col],palette = 'summer_r')
    plt.xlabel(f'Cancer {col}')
    plt.ylabel('Count')
    plt.title(f'Count of Cancer {col}')
    plt.show()
    
#Correlation matrix    
def plot_corr(data:pd.DataFrame)->None:
    corr = data.corr()
    plt.figure(figsize=(20,20))
    sns.heatmap(corr, cbar=True, square= True, fmt='.1f', annot=True, annot_kws={'size':15}, cmap='summer_r')
    plt.show()
    
#Distribution plot    
def plot_distribution(data:pd.DataFrame,col1:str,col2:str,col3:str)->None:
    plt.subplot(1,2,1)
    sns.FacetGrid(data, hue="diagnosis", height=6).map(sns.kdeplot, col1).add_legend()
    plt.subplot(1,2,2)
    sns.FacetGrid(data, hue="diagnosis", height=6).map(sns.kdeplot, col2).add_legend()
    plt.subplot(1,3,3)
    sns.FacetGrid(data, hue="diagnosis", height=6).map(sns.kdeplot, col3).add_legend()
    plt.show()
```

```python
plot_count(df,'diagnosis')
```

Most breast cancer cases were diagnosed as benign

```python
plot_corr(df)
```

```python
plot_distribution(df,'radius_mean','texture_mean','perimeter_mean')
```
