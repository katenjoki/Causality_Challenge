Data Visualization

```python
M = df[(df['diagnosis'] != 0)]
B = df[(df['diagnosis'] == 0)]

#Count plot
def plot_count(data:pd.DataFrame,col:str)->None:
    sns.countplot(data[col],palette = 'RdYlGn')
    plt.xlabel(f'Cancer {col}')
    plt.ylabel('Count')
    plt.title(f'Count of Cancer {col}')
    plt.show()
    
#Correlation matrix    
def plot_corr(data:pd.DataFrame)->None:
    corr = data.corr()
    plt.figure(figsize=(20,20))
    sns.heatmap(corr, cbar=True, square= True, fmt='.1f', annot=True, annot_kws={'size':15}, cmap='summer_r')
    plt.show()

#Scatter plot
def plot_scatter(df,col1,col2):
    plt.figure(figsize=(12, 7))
    sns.scatterplot(data = df, x=col1, y=col2, hue="diagnosis",palette='rainbow')
    plt.title(f'{col1} vs {col2}', size=16)
    plt.xticks(fontsize=14)
    plt.yticks( fontsize=14)
    plt.show()
    
#Distribution plot    
def plot_distribution(data:pd.DataFrame,col:str)->None:
    sns.FacetGrid(data, hue="diagnosis", height=6,palette='RdYlBu').map(sns.distplot, col).add_legend()
    plt.show()
```

```python
plot_count(df,'diagnosis')
```

Most breast cancer cases were diagnosed as benign

```python
plot_corr(df)
```

**Scatter plots of uncorrelated features** i.e correlation = 0

```python
plot_scatter(df,'compactness_mean','radius_se')
plot_scatter(df,'radius_se','fractal_dimension_worst')
```

**Scatter pots of some positively correlated features**

```python
plot_scatter(df,'radius_se','perimeter_se')
plot_scatter(df,'perimeter_mean','area_mean')
plot_scatter(df,'concavity_mean','compactness_mean')
```

**Scatter plots for negatively correlated features**

```python
plot_scatter(df,'radius_mean','fractal_dimension_se')
plot_scatter(df,'smoothness_se','area_mean')
```

**Features distribution, with hue=diagnosis**

```python
#some plots for radius
plot_distribution(df,'radius_mean')
plot_distribution(df,'radius_se')
plot_distribution(df,'radius_worst')
```

```python
#plots for texture
plot_distribution(df,'texture_mean')
plot_distribution(df,'texture_se')
plot_distribution(df,'texture_worst')
```

**Feature Scaling**

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler

le = LabelEncoder()
df.diagnosis = le.fit_transform(df.diagnosis)

cols = df.columns.to_list()
del cols[0]

scaled_features = df.copy()
features = scaled_features[cols]
scaler = StandardScaler()

scaled_features[cols] = scaler.fit_transform(features.values)
scaled_features.head()
```

```python
scaled_features.to_csv('../data/scaled_data.csv',index=False)
```
